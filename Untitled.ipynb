```python
#data PreProcessing
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, r2_score
import matplotlib.pyplot as plt

print("...............Crop Yield Prediction using Machine Learning...............")

# Load dataset
data = pd.read_csv("final_crop.csv")
print("Columns in dataset:", data.columns)

# Drop unwanted column if present
if 'Unnamed: 0' in data.columns:
    data = data.drop(columns=['Unnamed: 0'])

# Encode categorical columns
categorical_columns = ["Dist Name", "Crop"]
existing_categorical_columns = [col for col in categorical_columns if col in data.columns]

label_encoders = {}
for col in existing_categorical_columns:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
    label_encoders[col] = le

# Check for target column
if "Yield(Kg per ha)" not in data.columns:
    raise KeyError("❌ Target column 'Yield(Kg per ha)' not found in dataset!")

# Drop missing values
data = data.dropna()

# Split features and target
features = data.drop(columns=['Yield(Kg per ha)'])
target = data['Yield(Kg per ha)']

print("Feature columns:", features.columns.tolist())
print("Number of features:", len(features.columns))

# Scale features
scaler = MinMaxScaler()
features_scaled = scaler.fit_transform(features)

# Scale target
target = np.array(target).reshape(-1, 1)
scaler_target = MinMaxScaler()
target_scaled = scaler_target.fit_transform(target)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    features_scaled, target_scaled, test_size=0.2, random_state=42
)

# Convert to tensors
X_train = torch.tensor(X_train, dtype=torch.float32).unsqueeze(1)
y_train = torch.tensor(y_train, dtype=torch.float32)
X_test = torch.tensor(X_test, dtype=torch.float32).unsqueeze(1)
y_test = torch.tensor(y_test, dtype=torch.float32)

# Data loaders
train_loader = DataLoader(TensorDataset(X_train, y_train), batch_size=32, shuffle=True)
test_loader = DataLoader(TensorDataset(X_test, y_test), batch_size=32, shuffle=False)

print("✅ Preprocessing complete!")
```

```python
# CNN Model
import torch
import torch.nn as nn

class CNNModel(nn.Module):
    def __init__(self, input_dim, feature_length):
        super(CNNModel, self).__init__()
        self.conv1 = nn.Conv1d(in_channels=input_dim, out_channels=64, kernel_size=3, padding=1)
        self.relu = nn.ReLU()
        self.fc = nn.Linear(64 * feature_length, 1)

    def forward(self, x):
        x = self.conv1(x)
        x = self.relu(x)
        x = x.view(x.size(0), -1)
        x = self.fc(x)
        return x
```

```python
import torch
import torch.nn as nn

class HybridModel(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, feature_length):
        super(HybridModel, self).__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
        self.cnn = nn.Conv1d(in_channels=1, out_channels=64, kernel_size=3, padding=1)
        self.relu = nn.ReLU()
        self.fc_combined = nn.Linear(hidden_size + (64 * feature_length), 1)

    def forward(self, x):
        # LSTM branch
        lstm_out, _ = self.lstm(x)
        lstm_out = lstm_out[:, -1, :]  # Take output from the last time step

        # CNN branch
        cnn_out = self.cnn(x)
        cnn_out = self.relu(cnn_out)
        cnn_out = cnn_out.view(cnn_out.size(0), -1)

        # Combine features
        combined = torch.cat((lstm_out, cnn_out), dim=1)
        output = self.fc_combined(combined)
        return output
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

def train_model(model, train_loader, num_epochs=50):
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=0.001)

    for epoch in range(num_epochs):
        total_loss = 0
        for X_batch, y_batch in train_loader:
            optimizer.zero_grad()
            y_pred = model(X_batch)
            loss = criterion(y_pred, y_batch)
            loss.backward()
            optimizer.step()
            total_loss += loss.item()

        if epoch % 10 == 0:
            print(f"Epoch {epoch}, Loss: {total_loss / len(train_loader)}")
```

```python
import numpy as np
from sklearn.metrics import mean_absolute_error, r2_score

def evaluate_model(model, test_loader, scaler_target):
    model.eval()
    predictions, actuals = [], []

    with torch.no_grad():
        for X_batch, y_batch in test_loader:
            y_pred = model(X_batch)
            predictions.append(y_pred.detach().cpu().numpy())
            actuals.append(y_batch.detach().cpu().numpy())

    predictions = np.vstack(predictions)
    actuals = np.vstack(actuals)

    # Inverse transform back to original scale
    predictions = scaler_target.inverse_transform(predictions)
    actuals = scaler_target.inverse_transform(actuals)

    mse = np.mean((predictions - actuals) ** 2)
    mae = mean_absolute_error(actuals, predictions)
    r2 = r2_score(actuals, predictions)

    print(f"✅ Test MSE: {mse:.5f}")
    print(f"✅ Test MAE: {mae:.5f}")
    print(f"✅ R² Score: {r2:.5f}")

    return predictions, actuals


# Initialize models
cnn_model = CNNModel(input_dim=1, feature_length=X_train.shape[2])
hybrid_model = HybridModel(input_size=X_train.shape[2], hidden_size=50, num_layers=2, feature_length=X_train.shape[2])

# Train CNN
print("Training CNN...")
train_model(cnn_model, train_loader)

# Train Hybrid
print("Training Hybrid...")
train_model(hybrid_model, train_loader)
```

```python
#plotting the results
import matplotlib.pyplot as plt

def plot_results(predictions, actuals, title):
    plt.scatter(actuals, predictions, alpha=0.5)
    plt.plot([actuals.min(), actuals.max()], [actuals.min(), actuals.max()], 'r--', lw=2)
    plt.title(title)
    plt.xlabel('Actual Yield (Kg/ha)')
    plt.ylabel('Predicted Yield (Kg/ha)')
    plt.grid(True)
    plt.show()


# Evaluate CNN
print("\nCNN Evaluation:")
cnn_predictions, cnn_actuals = evaluate_model(cnn_model, test_loader, scaler_target)
plot_results(cnn_predictions, cnn_actuals, "CNN Predicted vs Actual")

# Evaluate Hybrid
print("\nHybrid Model Evaluation:")
hybrid_predictions, hybrid_actuals = evaluate_model(hybrid_model, test_loader, scaler_target)
plot_results(hybrid_predictions, hybrid_actuals, "Hybrid Predicted vs Actual")


def predict_yield(cnn_model, hybrid_model, new_data, scaler, scaler_target):
    # Ensure new_data is numpy array
    if isinstance(new_data, pd.DataFrame):
        new_data = new_data.values

    # Scale the input features
    new_data_scaled = scaler.transform(new_data)

    # Convert to tensor and reshape for models (same way as during training)
    new_data_tensor = torch.tensor(new_data_scaled, dtype=torch.float32).unsqueeze(1)  
    # Shape: [batch_size, 1, features]

    # Set models to evaluation mode
    cnn_model.eval()
    hybrid_model.eval()

    with torch.no_grad():
        # CNN Prediction
        cnn_pred_scaled = cnn_model(new_data_tensor)
        cnn_pred = scaler_target.inverse_transform(cnn_pred_scaled.detach().cpu().numpy())

        # Hybrid Prediction
        hybrid_pred_scaled = hybrid_model(new_data_tensor)
        hybrid_pred = scaler_target.inverse_transform(hybrid_pred_scaled.detach().cpu().numpy())

    # Prepare result as dictionary
    predictions = {
        "cnn_yield": cnn_pred.flatten(),
        "hybrid_yield": hybrid_pred.flatten()
    }
    return predictions


# Example new sample (update values as per your dataset columns)
new_sample = np.array([
    [2025, 5, 2, 15.5, 20.3, 325.7, 23.4]  # Adjust feature count/order to match training features
])

# Make prediction
predictions = predict_yield(cnn_model, hybrid_model, new_sample, scaler, scaler_target)
print(f"Predicted Yield (CNN): {predictions['cnn_yield'][0]:.2f} Kg/ha")
print(f"Predicted Yield (Hybrid): {predictions['hybrid_yield'][0]:.2f} Kg/ha")
```
